```python
def TrailerWrite(fileName):
    #Normal
    cap = cv2.VideoCapture(fileName)
    f = open(fileName[:-3]+"txt", "w") #note file endung must be 3 letters like mp4
    count = 0 
    countEnd = 0
    totalImage = []
    finalImage = []
    fakeCount = 0
    whilebool = True
    while(whilebool):

        count += 1 
        # Capture frame-by-frame
        ret, frame = cap.read()
        if not ret:
            whilebool=False
            break

        if count==1:
            totalImage = frame.copy().astype(np.uint32)
        elif np.count_nonzero(frame)!=0:
            countEnd+=1
            totalImage =np.add(totalImage,frame)
        if not ret:
            whilebool=False
            break
        # Our operations on the frame come here
        # Display the resulting frame
        if cv2.waitKey(1) & 0xFF == ord('q'):
            whilebool=False
            break

    print("total")

    
    finalImage = np.divide(totalImage.copy(), countEnd)
    print("final")
    stringFinal = np.array_str(finalImage, max_line_width=10000)
    f.write(stringFinal)
    



    # When everything done, release the capture
    cap.release()
    cv2.destroyAllWindows()
    #cv2.waitKey(0)
    np.reshape
    print(countEnd)
    return finalImage
```

```python
def ModeFinder(fileName):
    cap = cv2.VideoCapture(fileName)
    modeArr = []
    count = 0
    whilebool = True
    while(whilebool):

        # Capture frame-by-frame
        ret, frame = cap.read()
        if not ret:
            break
        count +=1
        print(count)
        modeArr.append(stats.mode(frame))

        if not ret:
            whilebool=False
            break
        # Our operations on the frame come here
        # Display the resulting frame
        if cv2.waitKey(1) & 0xFF == ord('q'):
            whilebool=False
            break
    return modeArr
```

```python
def TrailerReadTotal(fileName, genreName):
    f = open(fileName, "r")
    totalImage = []
    finalImage = []
    count = 0
    for line in f:
        image = TrailerWrite(line.strip()) #not sure if it cuts off \n or not
        if count==0:
            totalImage = image.copy().astype(np.float64)
        else:
            totalImage +=image
        count +=1
        print(count)

    finalImage = np.divide(totalImage.copy(), count)
    f = open(genreName+"txt", "w")
    stringFinal = np.array_str(finalImage, max_line_width=10000)
    f.write(stringFinal)
    plt.imshow(finalImage.astype(int))
```
