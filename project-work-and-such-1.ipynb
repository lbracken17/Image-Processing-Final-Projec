```python
def TrailerWrite(fileName):
    #This function agregates all the frames of a single trailer
    #Normal
    cap = cv2.VideoCapture(fileName)
    f = open(fileName[:-3]+"txt", "w") #note file endung must be 3 letters like mp4
    count = 0 
    countEnd = 0
    totalImage = []
    finalImage = []
    fakeCount = 0
    whilebool = True
    while(whilebool):
        count += 1 
        # Capture frame-by-frame
        ret, frame = cap.read()
        if not ret:
            whilebool=False
            break

        if count==1:
            totalImage = frame.copy().astype(np.uint32)
        elif np.count_nonzero(frame)!=0:
            countEnd+=1
            totalImage =np.add(totalImage,frame)
        if not ret:
            whilebool=False
            break
        # Our operations on the frame come here
        # Display the resulting frame
        if cv2.waitKey(1) & 0xFF == ord('q'):
            whilebool=False
            break

    print("total")

    
    finalImage = np.divide(totalImage.copy(), countEnd)
    print("final")
    stringFinal = np.array_str(finalImage, max_line_width=10000)
    f.write(stringFinal)
    



    # When everything done, release the capture
    cap.release()
    cv2.destroyAllWindows()
    #cv2.waitKey(0)
    np.reshape
    print(countEnd)
    return finalImage
```

```python
def EnhanceImage(image):
    #This function enhances final image agregates
    (cnt, binc) = ske.histogram(image, nbins=100)
    bincSort = np.sort(binc)
    bincShape = bincSort.shape

    print(bincShape)
    bincSize = bincShape[0]

    bincBottomThresh = bincSort[int(bincSize/33)] #20 was selected to allow for a 90% stretch
    bincTopThresh = bincSort[-int(bincSize/33)]
    print(bincBottomThresh)


    print(bincTopThresh)
    m = abs((255 - 0) / (bincTopThresh - bincBottomThresh))  # Slope of a line going through two points
    print(m)
    x = m * image
    newX = x.astype(int)
    
    newX[newX < 0] = 0  # Setting all negative values to 0
    newX[newX > 255] = 255  # Setting all saturated values to 1
    newX = newX.astype(int)
    
    plt.figure()
    plt.bar(bincSort, cnt, width=3)
    return newX
```
